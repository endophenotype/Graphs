```python
import networkx as nx
from networkx.algorithms import approximation
import math
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import pandas as pd
```

```python
ng = 100 # Число графов
nv = 100 # Число вершин в графе
l1 =  1/(nv*math.sqrt(nv))
list1 = nx.erdos_renyi_graph(nv, l1)
```

```python
nx.draw(list1, with_labels= True,)
plt.show()
```

```python
print("Cвязность узлов для графа:" )
print(nx.node_connectivity(list1))
print("Связность ребер графа:" )
print(nx.edge_connectivity(list1))
print("Транзитивность графа:" )
print(nx.transitivity(list1))
print("Кластерный коэффициент:" )
print(nx.average_clustering (list1))
print("Cловарь со значением кластеризации каждого узла:" )
print(nx.clustering(list1))
print("Cредний коэффициент кластеризации:" )
print(approximation.average_clustering(list1, trials=100))
```
